# 1/x

```python
%matplotlib inline
import matplotlib.pyplot as plt
x_d = [x for x in range(-100, 100) if x != 0]
y_d = [1 / x for x in x_d]
plt.clf()
plt.plot(x_d, y_d, linewidth=2, label="1/x")
plt.show()
```

# ln(x)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x_d = [x for x in range(0, 1000) if x != 0]
y_d = np.log(x_d)
plt.clf()
plt.plot(x_d, y_d, linewidth=2, label="ln(x)")
plt.show()
```

# e^x

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x_d = [x / 100 for x in range(0, 2000) if x != 0]
y_d = np.exp(x_d)
plt.clf()
plt.plot(x_d, y_d, linewidth=2, label="e^x")
plt.show()
```

# e^sin(x)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x_d = [x / 100 for x in range(0, 2000) if x != 0]
y_d = np.exp(np.sin(x_d))
plt.clf()
plt.plot(x_d, y_d, linewidth=2, label="e^x")
plt.show()
```
